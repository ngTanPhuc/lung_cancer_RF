```python
import pandas as pd
import os
```

```python
data_raw_path = os.path.join('..', 'data', 'raw', 'dataseter.csv')
df = pd.read_csv(data_raw_path)

df.head(10)
```

```python
df.info()
```

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python
df = df.drop_duplicates()  # Remove duplicate rows
df.rename(mapper={"GENDER": "IS_MALE"}, axis=1, inplace=True)  # Rename GENDER to IS_MALE
df["AGE"] = df["AGE"].astype("int16")  # Convert int64 to int 16
df.head(5)
```

```python
df.info()
```

```python
def change_dtype(val):
  val = str(val).lower()
  if val in ['m', 'yes']:
    return True
  elif val in ['f', 'no']:
    return False
  else:
    return val
```

```python
df["IS_MALE"] = df["IS_MALE"].map(change_dtype)
df["SMOKING"] = df["SMOKING"].map(change_dtype)
df["YELLOW_FINGERS"] = df["YELLOW_FINGERS"].map(change_dtype)
df["ANXIETY"] = df["ANXIETY"].map(change_dtype)
df["PEER_PRESSURE"] = df["PEER_PRESSURE"].map(change_dtype)
df["CHRONIC_DISEASE"] = df["CHRONIC_DISEASE"].map(change_dtype)
df["FATIGUE"] = df["FATIGUE"].map(change_dtype)
df["ALLERGY"] = df["ALLERGY"].map(change_dtype)
df["WHEEZING"] = df["WHEEZING"].map(change_dtype)
df["ALCOHOL_CONSUMING"] = df["ALCOHOL_CONSUMING"].map(change_dtype)
df["COUGHING"] = df["COUGHING"].map(change_dtype)
df["SHORTNESS_OF_BREATH"] = df["SHORTNESS_OF_BREATH"].map(change_dtype)
df["SWALLOWING_DIFFICULTY"] = df["SWALLOWING_DIFFICULTY"].map(change_dtype)
df["CHEST_PAIN"] = df["CHEST_PAIN"].map(change_dtype)
df["LUNG_CANCER"] = df["LUNG_CANCER"].map(change_dtype)
df.head(5)
```

```python
df.info()
```

```python
data_new_path = os.path.join('..', 'data', 'processed', 'dataset.csv')
df.to_csv(data_new_path, index=False)
```
